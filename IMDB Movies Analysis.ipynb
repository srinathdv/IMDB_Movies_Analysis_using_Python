## 20. Classify Movies Based on Ratings [Excellent, Good, and Average]

```python
#lets Create our own rating function
def rating(rat):
    if rat>=7.0:
        return "Excellent"
    elif rat>6.0:
        return "Good"
    else:
        return "Average"
```

```python
#add a new column
df['Rating_Category']=df['Rating'].apply(rating)
df
```

## 22. Find Unique Values From Genre 

```python
df.Genre
```

```python
#There are 1 to 3 Genres for ecah Movie
#Lets divide them to 2d and then 1d , then we can find the unique values

list2d=[]
for value in df['Genre']:
    list2d.append(value.split(','))
list2d  #Genre of each Movie
```

```python
list1d=[]
for value in list2d:   #subsets
    for value1 in value:  #takes one value from subset
        list1d.append(value1)
list1d
```

```python
#Now we have the genre list in 1d lets find unique genres count
unique = []
for value in list1d:
    if value not in unique:
        unique.append(value)
unique
```
